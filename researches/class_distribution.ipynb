#Распределение изображений по классам

```
import matplotlib.pyplot as plt
import gdown
import os
```

```
if not os.path.exists('train'):
    if not os.path.exists('train.zip'):
        train_zip_url = 'https://drive.google.com/uc?id=1QdhIxh1QUEuLgRb7DWa7RA7CA08ybNRJ'
        gdown.download(train_zip_url, 'train.zip', False)
    ! unzip train.zip -d .
    ! rm train.zip
```

```
# создаём список имён классов
class_names = ! ls train
[class_names] = class_names
class_names = class_names.split()
```

```
# определим количество изображений в каждом из классов
nums = []
for name in class_names:
    num = ! ls /content/train/$name | wc -l
    [num] = num
    nums.append(int(num))
```

```
plt.bar(class_names, nums)
plt.xticks(rotation=45)
_ = plt.title('Распределение изображений по эмоциям')
```

Видно, что мы имеем дело с несбалансированными данными.

В проекте выполнена аугментация каждой эмоции до максимальной (по количеству содержащихся в ней изображений). Процесс аугментации реализован в файле Augmentation.ipynb
