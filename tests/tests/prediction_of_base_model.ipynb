```
from keras_vggface.vggface import VGGFace
vgg = VGGFace()
```

Проверять инференс модели будем, например, на изображении Джулии Робертс.
Загрузим его из открытых источников.

```
! wget https://cdn.fw-daily.com/2019/10/51198314_2376181329082116_7009955984602172402_n-e1572276940494.jpg -O julia.jpg > /dev/null
```

Выполним предобработку изображения, инференс и декодирование согласно
документации, размещённой в проекте keras-vggface.git в файле README.md.

```
from keras_vggface import utils
from tensorflow.keras.preprocessing import image
img = image.load_img('julia.jpg', target_size=(224,224))
img = image.img_to_array(img)
img = img[None, ...]
print(f'dtype={img.dtype}  img.shape={img.shape}')
img = utils.preprocess_input(img)
preds = vgg.predict(img)
utils.decode_predictions(preds)
```

Как видно из результата декодирования, базовая модель работает корректно. С её точки зрения на фото изображена именно Джулия Робертс (вероятность 94%).
